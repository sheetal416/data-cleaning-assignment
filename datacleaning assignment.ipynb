```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'From_To':['LoNDon_paris','MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis','Brussels_londOn'],'FlightNumber':[10045, np.nan, 10065, np.nan, 10085],'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
# 1st sloution of filling missing values
fill_missing_values = df['FlightNumber'].interpolate().astype(int)
fill_missing_values
```

```python
#2nd solution spliting 2 cvolumns
split_two_words = df["From_To"].str.split('_', expand = True)
split_two_words.columns = ['From','To']
split_two_words
```

```python
#4th delete and attach new dataframe
df = df.drop('From_To',axis = 1)
df = df.join(split_two_words)
df
```

```python
# 5th delay and expand
recent_delays = df['RecentDelays'].apply(pd.Series)
recent_delays.columns = ['delay_{}'.format(s) for s in range(1,len(recent_delays.columns)+1)]
df = df.drop('RecentDelays',axis = 1).join(recent_delays)
df
```
